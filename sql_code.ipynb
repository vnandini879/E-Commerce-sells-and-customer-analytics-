## 3. Find the total sales per category.

```python
pd.read_sql("""
SELECT p."product category",
       SUM(oi.price) AS total_sales
FROM order_items oi
JOIN products p ON oi.product_id = p.product_id
GROUP BY p."product category"
ORDER BY total_sales DESC
""", conn)
```

4. Calculate the percentage of orders that were paid in installments.

```python
pd.read_sql("""
WITH OrderPaymentStatus AS (
    SELECT
        o.order_id,
        MAX(CASE WHEN p.payment_installments > 0 THEN 1 ELSE 0 END) AS has_installments_flag
    FROM
        orders o
    INNER JOIN
        payments p ON o.order_id = p.order_id
    GROUP BY
        o.order_id
)
SELECT
    SUM(has_installments_flag) AS num_orders_with_installments,
    COUNT(order_id) AS total_paid_orders,
    ROUND(
        (CAST(SUM(has_installments_flag) AS REAL) * 100.0) / COUNT(order_id), 2
    ) AS percentage_with_installments
FROM
    OrderPaymentStatus;
""", conn)
```

5. Count the number of customers from each state.

```python
pd.read_sql("""
SELECT
    customer_state,
    COUNT(customer_id) AS customer_count
FROM
    customers
GROUP BY
    customer_state
ORDER BY
    customer_count DESC;
""", conn)
```

# Intermediate Problems
1. Calculate the number of orders per month in 2018.

```python
pd.read_sql("""
SELECT
    strftime('%m', order_purchase_timestamp) AS month,
    COUNT(order_id) AS num_orders
FROM
    orders
WHERE
    strftime('%Y', order_purchase_timestamp) = '2018'
GROUP BY
    month
ORDER BY
    month;
""", conn)
```

2. Find the average number of products per order, grouped by customer city.

```python
pd.read_sql("""
SELECT
    c.customer_city,
    AVG(order_counts.num_products) AS avg_products_per_order
FROM
    customers c
JOIN
    orders o ON c.customer_id = o.customer_id
JOIN (
    SELECT
        order_id,
        COUNT(order_item_id) AS num_products
    FROM
        order_items
    GROUP BY
        order_id
) AS order_counts ON o.order_id = order_counts.order_id
GROUP BY
    c.customer_city
ORDER BY
    avg_products_per_order DESC
LIMIT 10;
""", conn)
```

3. Calculate the percentage of total revenue contributed by each product category.

```python
pd.read_sql("""
WITH CategoryRevenue AS (
    SELECT
        p."product category",
        SUM(oi.price) AS category_total_revenue
    FROM
        order_items oi
    JOIN
        products p ON oi.product_id = p.product_id
    GROUP BY
        p."product category"
),
TotalRevenue AS (
    SELECT
        SUM(price) AS overall_total_revenue
    FROM
        order_items
)
SELECT
    cr."product category",
    ROUND((cr.category_total_revenue * 100.0) / tr.overall_total_revenue, 2) AS percentage_of_total_revenue
FROM
    CategoryRevenue cr,
    TotalRevenue tr
ORDER BY
    percentage_of_total_revenue DESC
LIMIT 10;
""", conn)
```

4. Identify the correlation between product price and the number of times a product has been purchased.

```python
product_metrics_sql = pd.read_sql("""
WITH ProductPurchaseCounts AS (
    SELECT
        product_id,
        COUNT(order_item_id) AS purchase_count
    FROM
        order_items
    GROUP BY
        product_id
),
ProductAveragePrice AS (
    SELECT
        product_id,
        AVG(price) AS average_price
    FROM
        order_items
    GROUP BY
        product_id
)
SELECT
    ppc.product_id,
    ppc.purchase_count,
    pap.average_price
FROM
    ProductPurchaseCounts ppc
JOIN
    ProductAveragePrice pap ON ppc.product_id = pap.product_id;
""", conn)

correlation = product_metrics_sql['average_price'].corr(product_metrics_sql['purchase_count'])

print(f"Correlation between product price and number of times purchased: {correlation:.2f}")
```

5. Calculate the total revenue generated by each seller, and rank them by revenue.

```python
pd.read_sql("""
SELECT
    s.seller_id,
    SUM(oi.price) AS total_revenue
FROM
    order_items oi
JOIN
    sellers s ON oi.seller_id = s.seller_id
GROUP BY
    s.seller_id
ORDER BY
    total_revenue DESC
LIMIT 10;
""", conn)
```

# Advanced Problems
1. Calculate the moving average of order values for each customer over their order history.

```python
pd.read_sql("""
WITH CustomerOrderValues AS (
    SELECT
        o.customer_id,
        o.order_purchase_timestamp,
        p.payment_value
    FROM
        orders o
    LEFT JOIN
        payments p ON o.order_id = p.order_id
)
SELECT
    customer_id,
    order_purchase_timestamp,
    payment_value,
    AVG(payment_value) OVER (
        PARTITION BY customer_id
        ORDER BY order_purchase_timestamp
        ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
    ) AS moving_avg_order_value
FROM
    CustomerOrderValues
ORDER BY
    customer_id,
    order_purchase_timestamp
LIMIT 10;
""", conn)
```

2. Calculate the cumulative sales per month for each year.

```python
pd.read_sql("""
WITH MonthlySales AS (
    SELECT
        strftime('%Y', o.order_purchase_timestamp) AS year,
        strftime('%m', o.order_purchase_timestamp) AS month,
        SUM(p.payment_value) AS monthly_total_sales
    FROM
        orders o
    LEFT JOIN
        payments p ON o.order_id = p.order_id
    GROUP BY
        year, month
)
SELECT
    year,
    month,
    monthly_total_sales,
    SUM(monthly_total_sales) OVER (
        PARTITION BY year
        ORDER BY month
    ) AS cumulative_sales
FROM
    MonthlySales
ORDER BY
    year, month
LIMIT 20;
""", conn)
```

3. Calculate the year-over-year growth rate of total sales.

```python
pd.read_sql("""
WITH YearlySales AS (
    SELECT
        strftime('%Y', o.order_purchase_timestamp) AS year,
        SUM(p.payment_value) AS total_sales
    FROM
        orders o
    LEFT JOIN
        payments p ON o.order_id = p.order_id
    GROUP BY
        year
)
SELECT
    year,
    total_sales,
    ROUND(
        ((total_sales - LAG(total_sales, 1) OVER (ORDER BY year)) * 100.0) / LAG(total_sales, 1) OVER (ORDER BY year), 2
    ) AS yoy_growth_rate
FROM
    YearlySales
ORDER BY
    year;
""", conn)
```

4. Calculate the retention rate of customers, defined as the percentage of customers who make another purchase within 6 months of their first purchase.

```python
pd.read_sql("""
WITH CustomerFirstPurchases AS (
    SELECT
        c.customer_unique_id,
        MIN(o.order_purchase_timestamp) AS first_purchase_date
    FROM
        orders o
    JOIN
        customers c ON o.customer_id = c.customer_id
    GROUP BY
        c.customer_unique_id
),
RetainedUniqueCustomers AS (
    SELECT DISTINCT
        cf.customer_unique_id
    FROM
        orders o
    JOIN
        customers c ON o.customer_id = c.customer_id
    JOIN
        CustomerFirstPurchases cf ON c.customer_unique_id = cf.customer_unique_id
    WHERE
        (julianday(o.order_purchase_timestamp) - julianday(cf.first_purchase_date)) > 0
        AND (julianday(o.order_purchase_timestamp) - julianday(cf.first_purchase_date)) <= 180
)
SELECT
    COUNT(DISTINCT ruc.customer_unique_id) AS num_retained_customers,
    (SELECT COUNT(DISTINCT customer_unique_id) FROM customers) AS total_unique_customers,
    ROUND(
        (CAST(COUNT(DISTINCT ruc.customer_unique_id) AS REAL) * 100.0) / (SELECT COUNT(DISTINCT customer_unique_id) FROM customers),
        2
    ) AS retention_rate_percentage
FROM
    RetainedUniqueCustomers ruc;
""", conn)
```

5. Identify the top 3 customers who spent the most money in each year.

```python
pd.read_sql("""
WITH CustomerYearlySpend AS (
    SELECT
        strftime('%Y', o.order_purchase_timestamp) AS year,
        o.customer_id,
        SUM(p.payment_value) AS total_spend
    FROM
        orders o
    JOIN
        payments p ON o.order_id = p.order_id
    GROUP BY
        year, o.customer_id
),
RankedCustomerSpend AS (
    SELECT
        year,
        customer_id,
        total_spend,
        ROW_NUMBER() OVER (PARTITION BY year ORDER BY total_spend DESC) AS customer_rank
    FROM
        CustomerYearlySpend
)
SELECT
    year,
    customer_id,
    total_spend
FROM
    RankedCustomerSpend
WHERE
    customer_rank <= 3
ORDER BY
    year ASC, customer_rank ASC;
""", conn)
```
